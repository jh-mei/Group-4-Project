### How does weather affect the West Nile Virus?

Weather factors are important drivers of WNV transmission due to their direct effect in mosquito biology.

* According to [study drivers of West Nile virus human illness in the Chicago](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0227160)
    * higher than normal average annual temperatures are associated with an increased likelihood of higher WNV disease incidence
    * higher than normal precipitation was positively associated with higher human WNV cases in the eastern region of the United States, but this relationship was reversed for the western region
    * Another study identified drought as an important driver of WNV epidemics in the United States
    * hot weather conditions, warm winters, and higher MIR in earlier weeks increased the probability of an area of having a WNV human case.

#### Missing Tavg

We will replace the missing 'Tavg' data with the average of the Tmin and Tmax of that day.

```python
#Suggestions to replace missing Tavg with average of Tmin and Tmax 

def missing_Tavg(row):
    if row['Tavg'] == 'M':
        row['Tavg'] = (row['Tmax'] + row['Tmin']) / 2
    return row['Tavg']
```

```python
weather['Tavg'] = weather.apply(missing_Tavg, axis=1)
weather['Tavg'] = weather['Tavg'].astype(int)
```

## Merge Train and Weather datasets

### Assigning stations to train and test data points

Before we merge the two datasets, we have decided to assign weather data based on the trap's proximity to the nearest weather station. This should provide us with a more accurate representation of the weather condtions for each trap.

```python
def closest_station(latitude, longitude):
    '''Function to deteremine which weather station's data to use based on station proximity to trap'''
                    
    station_1_coord = (41.995, -87.933)
    station_2_coord = (41.786, -87.752)
    entry_coord = (latitude, longitude)
    
    if geodesic(entry_coord, station_1_coord) < geodesic(entry_coord, station_2_coord):
        return 1
    else:
        return 2
```

```python
#Include new 'station' column in train dataset

train['Station'] = train.apply(lambda x: closest_station(x['Latitude'], x['Longitude']), axis=1)
```

```python
#Include new 'station' column in test dataset

test['Station'] = test.apply(lambda x: closest_station(x['Latitude'], x['Longitude']), axis=1)
```

#### Imputing new addresses in test (not in train)

There are 13 new traps in the test set that is not in the train dataset. In total, these 13 traps give ~10k entries which is around 1% of the test dataset. 

Our group felt that it might be better to assign them a 'risk_rating' based on the closest trap available from the train dataset. As per our map above, we see that WnvPresent is usually in clusters, so it would be a reasonable assumption that the new trap in the test dataset should likely have a similar 'risk_rating' as a nearby trap.

```python
#138 unique addresses of traps in train dataset

unique_address = train_weather.groupby(['Address', 'risk_rating'])[['Latitude','Longitude']].mean().dropna()
unique_address = unique_address.reset_index()
unique_address_coords = list(zip(unique_address['Latitude'], unique_address['Longitude']))
```

```python
#These are the 13 coordinates not in the train set but are in the test set

coords_train = list(zip(train_weather['Latitude'],train_weather['Longitude']))
coords_test = list(zip(test_weather['Latitude'],test_weather['Longitude']))
coords_diff= list(set(coords_test)-set(coords_train))

coords_diff
```

```python
def imput_risk_rating(coords_diff): 
    '''function to find the nearest trap from the train dataset'''
    
    nearest = []
    for i in range(len(coords_diff)):
        lat_lon = []
        distance = geodesic(coords_diff[i], unique_address_coords[0])
        for j in range(len(unique_address_coords)):
            if geodesic(coords_diff[i], unique_address_coords[j]) < distance:
                distance = geodesic(coords_diff[i], unique_address_coords[j])
                lat_lon= unique_address_coords[j]
        nearest.append(lat_lon)
    return nearest
        
```

```python
#These are the coordinates of the 13 closest trap in the train dataset

imput_from_train = imput_risk_rating(coords_diff)
imput_from_train 
```

```python
#Update the 13 'new' traps with the 'risk_rating' of its closest trap in the train dataset

ratings=[]
for i in range(len(imput_from_train)):
    ratings.append(unique_address.loc[(round(unique_address['Latitude'],4)== round(imput_from_train[i][0],4)) & (round(unique_address['Longitude'],4)== round(imput_from_train[i][1],4)), 'risk_rating'].values[0])

reference = list(zip(coords_diff,ratings))
reference_df = pd.DataFrame(reference, columns=['coord','risk'])
reference_df
```

```python
#Apply on the entries with these 13 traps (they are currently null)

test_weather['risk_rating'] = test_weather.apply(
    lambda x: reference_df.loc[
    (reference_df['coord']==(x['Latitude'], x['Longitude'])),'risk'].values[0]
    if(pd.isnull(x['risk_rating'])) else x['risk_rating'], axis = 1)
```
