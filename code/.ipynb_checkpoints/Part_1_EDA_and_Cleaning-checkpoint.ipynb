## Introduction

### Background

In the United States, West Nile Virus (WNV) is the leading cause of mosquito-borne disease. It spreads to humans from the bite of an infected mosquito, and usually starts in summer. Since there are no vaccines to prevent or medicines to treat the virus, spraying by pesticides is one of the methods adopted to reduce the risk. WNV results in significant medical impact - 1 in 5 infected persons develop a fever, and 1 in 150 infected persons develop serious illness. [Source: CDC](https://www.cdc.gov/westnile/index.html)

We are a team of data scientists at the Disease And Treatment Agency. In the last few years, there have been several episodes of the West Nile Virus (WNV) outbreak in Chicago. The aim of our project is to leverage on weather, location and time data to predict West Nile Virus occurrences in Chicago. ### How does weather affect the West Nile Virus?

Weather factors are important drivers of WNV transmission due to their direct effect in mosquito biology.

* According to [study drivers of West Nile virus human illness in the Chicago](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0227160)
    * higher than normal average annual temperatures are associated with an increased likelihood of higher WNV disease incidence
    * higher than normal precipitation was positively associated with higher human WNV cases in the eastern region of the United States, but this relationship was reversed for the western region
    * Another study identified drought as an important driver of WNV epidemics in the United States
    * hot weather conditions, warm winters, and higher MIR in earlier weeks increased the probability of an area of having a WNV human case.

### Lagged Weather Features

According to the CDC, the Culex species of mosquitos that transmit West Nile Virus (WNV) take approximately 7-10 days to develop from eggs to full adult mosquitos. [Source](https://www.cdc.gov/mosquitoes/about/life-cycles/culex.html). 

Considering that transmission starts from these mosquitos feeding on dead birds and then passing on the virus to humans, we have decided to **lag our weather features by 14 days**. This should be a better reflection of whether prevailing weather conditions a fortnight ago will result in increase in WNV occurrences. 

```python
#Renamed 'Date' columns to 'Date Original'

weather = weather.rename(columns={'Date':'Date_Original'})
```

```python
#Add a new 'Date' column to for the 14 day lag - the train dataset will then be matched by this new 'Date' column

weather['Date'] = weather['Date_Original'] + pd.to_timedelta(14, unit='d')
```

#### Missing Tavg

We will replace the missing 'Tavg' data with the average of the Tmin and Tmax of that day.

```python
#Suggestions to replace missing Tavg with average of Tmin and Tmax 

def missing_Tavg(row):
    if row['Tavg'] == 'M':
        row['Tavg'] = (row['Tmax'] + row['Tmin']) / 2
    return row['Tavg']
```

```python
weather['Tavg'] = weather.apply(missing_Tavg, axis=1)
weather['Tavg'] = weather['Tavg'].astype(int)
```

## Merge Train and Weather datasets

### Assigning stations to train and test data points

Before we merge the two datasets, we have decided to assign weather data based on the trap's proximity to the nearest weather station. This should provide us with a more accurate representation of the weather condtions for each trap.

```python
def closest_station(latitude, longitude):
    '''Function to deteremine which weather station's data to use based on station proximity to trap'''
                    
    station_1_coord = (41.995, -87.933)
    station_2_coord = (41.786, -87.752)
    entry_coord = (latitude, longitude)
    
    if geodesic(entry_coord, station_1_coord) < geodesic(entry_coord, station_2_coord):
        return 1
    else:
        return 2
```

```python
#Include new 'station' column in train dataset

train['Station'] = train.apply(lambda x: closest_station(x['Latitude'], x['Longitude']), axis=1)
```

```python
#Include new 'station' column in test dataset

test['Station'] = test.apply(lambda x: closest_station(x['Latitude'], x['Longitude']), axis=1)
```

#### Imputing new addresses in test (not in train)

There are 13 new traps in the test set that is not in the train dataset. In total, these 13 traps give ~10k entries which is around 1% of the test dataset. 

Our group felt that it might be better to assign them a 'risk_rating' based on the closest trap available from the train dataset. As per our map above, we see that WnvPresent is usually in clusters, so it would be a reasonable assumption that the new trap in the test dataset should likely have a similar 'risk_rating' as a nearby trap.

```python
#138 unique addresses of traps in train dataset

unique_address = train_weather.groupby(['Address', 'risk_rating'])[['Latitude','Longitude']].mean().dropna()
unique_address = unique_address.reset_index()
unique_address_coords = list(zip(unique_address['Latitude'], unique_address['Longitude']))
```

```python
#These are the 13 coordinates not in the train set but are in the test set

coords_train = list(zip(train_weather['Latitude'],train_weather['Longitude']))
coords_test = list(zip(test_weather['Latitude'],test_weather['Longitude']))
coords_diff= list(set(coords_test)-set(coords_train))

coords_diff
```

```python
def imput_risk_rating(coords_diff): 
    '''function to find the nearest trap from the train dataset'''
    
    nearest = []
    for i in range(len(coords_diff)):
        lat_lon = []
        distance = geodesic(coords_diff[i], unique_address_coords[0])
        for j in range(len(unique_address_coords)):
            if geodesic(coords_diff[i], unique_address_coords[j]) < distance:
                distance = geodesic(coords_diff[i], unique_address_coords[j])
                lat_lon= unique_address_coords[j]
        nearest.append(lat_lon)
    return nearest
        
```

```python
#These are the coordinates of the 13 closest trap in the train dataset

imput_from_train = imput_risk_rating(coords_diff)
imput_from_train 
```

```python
#Update the 13 'new' traps with the 'risk_rating' of its closest trap in the train dataset

ratings=[]
for i in range(len(imput_from_train)):
    ratings.append(unique_address.loc[(round(unique_address['Latitude'],4)== round(imput_from_train[i][0],4)) & (round(unique_address['Longitude'],4)== round(imput_from_train[i][1],4)), 'risk_rating'].values[0])

reference = list(zip(coords_diff,ratings))
reference_df = pd.DataFrame(reference, columns=['coord','risk'])
reference_df
```

```python
#Apply on the entries with these 13 traps (they are currently null)

test_weather['risk_rating'] = test_weather.apply(
    lambda x: reference_df.loc[
    (reference_df['coord']==(x['Latitude'], x['Longitude'])),'risk'].values[0]
    if(pd.isnull(x['risk_rating'])) else x['risk_rating'], axis = 1)
```

### Relative Humidity

The relationship between humidity and WNV occurrences is ambiguous [Source](https://www.medicalnewstoday.com/articles/west-nile-virus-in-the-us-a-case-study-on-climate-change-and-health). On one hand, high humidity, which is usually a result of high precipitation, play a role in providing the conditions necessary for mosquitos to breed. However, some research have also argued that droughts and dry weather might actually cause mosquitos to migrate further as they look for new breeding grounds, therefore causing wider spread of the virus.

Our group believed that relative humidity might be an interesting feature to explore and it can be calculated from temperature and dewpoint temperature. Therefore, rather than manually creating an interaction term of Tavg x DewPoint, we have decided to create a 'relative_humidity' feature. The formulation of relative humidity (RH) is as below. 𝑒 is the water vapor pressure in the air and 𝑒𝑠 is the saturated water vapor pressure. 

While there are several mathematical formulations to working out relative humidity, we credit [this discussion](https://earthscience.stackexchange.com/questions/20577/relative-humidity-approximation-from-dew-point-and-temperature) for providing a simple formula from where we adapted our python calculations.

$ RH=100\times \frac{e}{e_s(T)} $

```python
#Convert Farenheit to Celsius 

train_weather['Tavg_c'] = ((train_weather['Tavg']-32) * 5/9)
train_weather['DewPoint_c'] = ((train_weather['DewPoint']-32) * 5/9)

test_weather['Tavg_c'] = ((test_weather['Tavg']-32) * 5/9)
test_weather['DewPoint_c'] = ((test_weather['DewPoint']-32) * 5/9)
```

```python
#calculate amount of water vapor in air
#calculate saturated vapour pressure
 
train_weather['Vap_Pressure_Air'] = np.exp((17.625 * train_weather['DewPoint_c']) / (train_weather['DewPoint_c'] + 243.04))
train_weather['Vap_Pressure_Saturated'] = np.exp((17.625 * train_weather['Tavg_c'] ) / (train_weather['Tavg_c'] + 243.04))

test_weather['Vap_Pressure_Air'] = np.exp((17.625 * test_weather['DewPoint_c']) / (test_weather['DewPoint_c'] + 243.04))
test_weather['Vap_Pressure_Saturated'] = np.exp((17.625 * test_weather['Tavg_c'] ) / (test_weather['Tavg_c'] + 243.04))
```

```python
train_weather['relative_humidity'] = (train_weather['Vap_Pressure_Air'] /train_weather['Vap_Pressure_Saturated'])*100
test_weather['relative_humidity'] = (test_weather['Vap_Pressure_Air'] /test_weather['Vap_Pressure_Saturated'])*100
```

```python
#Drop the Vapour Pressure and Tavg_c/Dewpoint_c columns created to calculate relative humidity

train_weather.drop(columns=['Vap_Pressure_Air','Vap_Pressure_Saturated', 'Tavg_c', 'DewPoint_c'], inplace=True)
test_weather.drop(columns=['Vap_Pressure_Air','Vap_Pressure_Saturated', 'Tavg_c', 'DewPoint_c'], inplace=True)
```

### Presence of WMV after within a week of Spraying

In this segment we will study the effectiveness of spray approximately 1 - 7 days after spraying.  We will locate the presence of WMV at the specific points prior to spray, and will evaluate if the presence of WMV remains after spray.

```python
# Create function to add year, month and day columns
def addymdcols(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
```

```python
## Create function to obtain the dates a day before and after spray
def get_date(df, year, month, afterday, beforeday):
    return df.loc[(df['year']==year)&(df['month']==month)&(df['day']>=afterday)&(df['day']<beforeday)]
```

```python
## Create function to plot spray against presence of WMV 
## before and after spray
def plotmap_spraytrain(spray, trainbefore, trainafter, date):

    fig, ax = plt.subplots(figsize=(15,15))

    chicago.plot(ax = ax, alpha=0.4, color='gray')
    spray.plot(ax = ax, color = 'yellow', marker = 's', label = 'Spray')
    trainbefore[trainbefore['WnvPresent']== 1].plot(ax = ax, color = 'blue', marker = 'o', markersize=100, label = 'Before')
    trainafter[trainafter['WnvPresent']== 1].plot(ax = ax, color = 'red', marker = 'o', label = 'After')
    plt.title(f'Effects of spraying 1 week after {date}')
    plt.legend();
```
